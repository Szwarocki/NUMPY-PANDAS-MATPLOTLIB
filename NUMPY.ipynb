```python
A = np.array([[1, 1], [2, 2], [3, 3]])
print(A, '\n')
print(np.insert(A, 1, 5,axis =None), '\n') # spłąszcza i wstawia element 5 na indeks 1
print(np.insert(A, 1, 5,axis =1), '\n') #dodaje kolumne "5" na na nową kolumne która będzie indeksem jeden
A
print(np.insert(A, 1, 5, axis=0), '\n')#nowy wierszz 
```

```python
A = np.arange(20).reshape(5,4)
print(A, '\n')

print(np.delete(A, 1, axis=0), '\n')  #usuwa 1 wiersz
print(A, '\n')
print(np.delete(A, [1, 2], 1), '\n') # usuwa kolumny o indeksaach 1 i 2 nie 1 do 2
print(np.delete(A, [1, 2, 5, 7, 9, 12, 17, 19])) # spłaszcza i usuwa elemnty
```

```python
A = np.arange(5, 20).reshape((5, 3))
B = np.arange(15).reshape((5, 3))

print(A < B, '\n')
print(A == A, '\n')
print(A == 17, '\n')
print(np.any( A > 17 ), '\n') # prawdziwe, jeśli każdy element z osobna spełnia warunek 
print(np.all( A > 17 )) # prawdziwe jeśli wszystkie elementy spełniają warunek
```

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])

print(x, '\n')
print(y, '\n')

# łączenie wierszowe
print(np.concatenate((x, y), axis=0), '\n')
print(np.vstack((x,y)), '\n')

# łączenie kolumnowe
print(np.concatenate((x,y.T), axis=1), '\n')
print(np.hstack((x,y.T)))
```

```python
x = np.arange(9)
print(x, '\n')
print(np.split(x, 3), '\n') # podział na 3 podtablice
print(np.split(x, [2, 4, 7]), '\n') # podział na 4 podtablice o 
                                # wskazanych początkach

x = x.reshape((3, 3))
print(x, '\n')
print(np.split(x, 3, axis=1), '\n')

# podział ze względu na kolumny
x = np.arange(16.0).reshape(4, 4)
print(x, '\n')
print(np.hsplit(x, 2), '\n')
print(np.hsplit(x, [1, 3]), '\n')

# podział ze względu na wiersze
print(np.vsplit(x, 2), '\n')
print(np.vsplit(x, [1, 3]), '\n')
```

```python
import numpy as np

# Tworzymy macierz 2x3
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Oryginalna macierz:\n", matrix)
print("Kształt oryginalnej macierzy:", matrix.shape,'\n')  # Wynik: (2, 3)

new_matrix_1 = matrix[np.newaxis, :, :]
print("Nowa macierz (1x2x3):\n", new_matrix_1)
print("Kształt nowej macierzy:", new_matrix_1.shape,'\n')  # Wynik: (1, 2, 3)

new_matrix_2 = matrix[:, np.newaxis, :]
print("Nowa macierz (2x1x3):\n", new_matrix_2)
print("Kształt nowej macierzy:", new_matrix_2.shape,'\n')  # Wynik: (2, 1, 3)

```
