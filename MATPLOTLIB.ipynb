```python
import matplotlib.pyplot as plt

# Dane dla wykresu koowego Pac-Mana
colors = ['yellow', 'white']
x = [80, 20]

# Tworzenie wykresu koowego Pac-Mana
plt.pie(x, colors=colors, startangle=45)

# Dodawanie okrgu (oka) do Pac-Mana po narysowaniu kawakw wykresu
eye_x = 0.4  # Wsprzdna X oka
eye_y = 0.6  # Wsprzdna Y oka
eye_radius = 0.1  # Promie oka

eye_circle = plt.Circle((eye_x, eye_y), eye_radius, color='black')
plt.gca().add_patch(eye_circle)

# Wywietlanie wykresu
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Tworzenie danych do wykresu 3D
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y) #stworzenie siatki punktw na ktrych podstawie zostawnie wygenerowany wykres 3D
Z = np.sin(np.sqrt(X**2 + Y**2))

# Tworzenie wykresu 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Dodawanie etykiet
ax.set_xlabel('O X')
ax.set_ylabel('O Y')
ax.set_zlabel('O Z')
ax.set_title('Wykres 3D')

# Dodawanie paska kolorw
fig.colorbar(surf)

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Tworzenie figury i osi 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Wsprzdne punktu
x = 1
y = 2
z = 3

# Rysowanie punktu
ax.scatter(x, y, z, c='r', marker='o')

# Etykiety osi
ax.set_xlabel('O X')
ax.set_ylabel('O Y')
ax.set_zlabel('O Z')

# Tytu
ax.set_title('Trjwymiarowy Punkt')

# Wywietlenie wykresu
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dane dla pierwszego wykresu
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

# Dane dla drugiego wykresu
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)

# Tworzenie rysunku i osi
plt.figure()

# Rysowanie pierwszego wykresu liniowego (sinus)
plt.plot(x1, y1, label='Sin(x)', color='blue', linestyle='-', marker='x')

# Rysowanie drugiego wykresu liniowego (cosinus)
plt.plot(x2, y2, label='Cos(x)', color='red', linestyle='--')

# Dodawanie etykiet
plt.xlabel('O X')
plt.ylabel('O Y')
plt.title('Dwa Wykresy Liniowe')

# Dodawanie legendy
plt.legend()

# Wywietlanie wykresu
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Przykadowe dane w postaci DataFrame
data = {
    'x': [1, 2, 3, 4, 5],
    'y1': [10, 14, 8, 16, 12],
    'y2': [5, 8, 6, 9, 7]
}

df = pd.DataFrame(data)

# Tworzenie rysunku i osi
plt.figure()

# Rysowanie pierwszego wykresu liniowego (y1)
plt.plot(df['x'], df['y1'], label='Wykres y1', color='blue', linestyle='-', marker='o')

# Rysowanie drugiego wykresu liniowego (y2)
plt.plot(df['x'], df['y2'], label='Wykres y2', color='red', linestyle='--', marker='x')

# Dodawanie etykiet
plt.xlabel('O X')
plt.ylabel('O Y')
plt.title('Dwa Wykresy Liniowe na Podstawie DataFrame')

# Dodawanie legendy
plt.legend()

# Wywietlanie wykresu
plt.show()
```
